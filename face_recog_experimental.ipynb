```python
img.shape #displays the size of the image vector
gray_img = cv.cvtColor(img,cv.COLOR_BGR2GRAY) #BGR color space to gray
```

```python
cv2_imshow(gray_img)
```

```python
gray_img.shape
```

```python
face_classifier = cv.CascadeClassifier(
    cv.data.haarcascades + "haarcascade_frontalface_default.xml"
)
#Cascade Classifier is a model (using softmax ig)
#we fit harcascade ( a face dataset ) into the model
```

```python
face = face_classifier.detectMultiScale(
    gray_img, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
#gray_image
#scaleFactor = image size reduction, 10 means reduce by 110-100%
#minneighbors = distance between adjacent rectangles (accuracy)
#minSize = determines the minimum size of faces (accuracy)
```

```python
for (x,y,l,h) in face: #for each face in picture
  cv.rectangle(img, (x,y),(x+l, y+h),(0,0,255),2) #fit rectangle onto each face
  #drawRectangle(image object, first two coordinates, new two coordinates, color in RGB, size of border)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB) #BGR color space to RGB
```
